```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
import numpy as np
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt


# create the training & test sets, skipping the header row with [1:]
dataset = pd.read_csv("train.csv")
target = dataset[[0]].values.ravel()[:100]
train = dataset.iloc[:,1:].values[:100]
f = open("train.csv")
header = f.readline().split(",")
f.close()
test = pd.read_csv("test.csv").values

print("Hey")
# create and train the random forest
# multi-core CPUs can use: rf = RandomForestClassifier(n_estimators=100, n_jobs=2)
rf = RandomForestClassifier(n_estimators=5)
rf.fit(train, target)


print(dir(rf))
#np.savetxt('submission_rand_forest.csv', np.c_[range(1,len(test)+1),pred], delimiter=',', header = 'ImageId,Label', comments = '', fmt='%d')


'''

for dectree in rf.estimators_:
    print(dectree.tree_.feature,dectree.tree_.threshold,dectree.tree_.children_left,dectree.tree_.children_right)
    
def get_code(tree, feature_names):
        left      = tree.tree_.children_left
        right     = tree.tree_.children_right
        threshold = tree.tree_.threshold
        features  = [feature_names[i] for i in tree.tree_.feature]
        value = tree.tree_.value

        def recurse(left, right, threshold, features, node):
                if (threshold[node] != -2):
                        print("if ( " + features[node] + " <= " + str(threshold[node]) + " ) {")
                        if left[node] != -1:
                                recurse (left, right, threshold, features,left[node])
                        print("} else {")
                        if right[node] != -1:
                                recurse (left, right, threshold, features,right[node])
                        print("}")
                else:
                        print("return " + str(value[node]))

        recurse(left, right, threshold, features, 0)

get_code(rf.estimators_[0],header)

'''


dataset = pd.read_csv("train.csv")
target = dataset[[0]].values.ravel()[10000:20000]
train = dataset.iloc[:,1:].values[10000:20000]

imgs = ["numbers_thick\\"+str(i)+"_thick.png" for i in range(0,10)]

mat = []
for img in imgs:
    img = Image.open(img)
    imggray = img.convert('LA')
    plt.figure(figsize=(9, 6))
    plt.imshow(imggray)
    imgmat = np.array(list(imggray.getdata(band=0)), int)
    imgmat = np.array([int(255 - i) for i in imgmat])
    mat.append(imgmat)
    
mat = np.array(mat)

rf = RandomForestClassifier(n_estimators=20)
df = tree.DecisionTreeClassifier(min_samples_leaf=1)
df.fit(train, target)
#pred = rf.predict(mat)
#print(pred)

# create and train the random forest
# multi-core CPUs can use: rf = RandomForestClassifier(n_estimators=100, n_jobs=2)
#rf = RandomForestClassifier(n_estimators=5)
#rf.fit(train, target)



for dectree in rf.estimators_:
    print(dectree.tree_.feature,dectree.tree_.threshold,dectree.tree_.children_left,dectree.tree_.children_right)
    
def binaryarr(num,bits):
    for item in range(0,len(num[0])):
        if(num[0][item] !=0):
            return(str(bits) + "\'b" +("{0:0" +str(bits)+"b}").format(item))
        
def binary(num,bits):
            return(str(bits) + "\'b" +("{0:0" +str(bits)+"b}").format(num))

def formatbin(feature,comparison,thresh):
    feat = feature[:5]
    index = "["+binary(int(feature[5:]),12)+"]"
    comp = comparison
    if(not thresh.is_integer()):
        thresh+=0.5
        comp = "<"
    binthresh = binary(int(thresh),8)
    
    return feat + index + comp + binthresh;
      
def get_code(tree, feature_names):
        writestring = [] 
        left      = tree.tree_.children_left
        right     = tree.tree_.children_right
        threshold = tree.tree_.threshold
        features  = [feature_names[i] for i in tree.tree_.feature]
        value = tree.tree_.value

        def recurse(left, right, threshold, features, node,writestring):
                if (threshold[node] != -2):
                        writestring.append(" (" + formatbin(features[node],"<=",threshold[node]) + ") ?")
                        #print("  (" + formatbin(features[node],"<=",threshold[node]) + ") ?")
                        if left[node] != -1:
                                recurse (left, right, threshold, features,left[node],writestring)
                        writestring.append(":")
                        #print(":")
                        if right[node] != -1:
                                recurse (left, right, threshold, features,right[node],writestring)
                        writestring.append(" \n")
                        #print(" \n")
                else:
                        writestring.append(str(binaryarr(value[node],8)) + "\n")
                        #print(str(binaryarr(value[node],8)) + "\n")
        recurse(left, right, threshold, features, 0,writestring)
        return "".join(writestring)

s = get_code(df,header)
print(s)
f = open("verilogdt.v","w")
f.write(s)
f.close()

pred = df.predict(mat)
print(pred)
```
